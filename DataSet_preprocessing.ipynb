```python
def generate_genres(g):
    genres=[]
    for i in g:
        genre=i.get("name")
        if genre=='Science Fiction':
            genres.append('Sci-Fi')
        else:
            genres.append(genre)
        if genres==[]:
            return np.nan
        else:
            return " ".join(genres)
            
    
```

```python
df_2017['genres']=df_2017['genres'].apply(generate_genres)
```

```python
def get_cast(x, actor_index=None, director=None):
    if director:
        directo=[]
        for i in x:
            if i.get('job') == "Director":
                directo.append(i.get("name"))
        if directo == []:
            return np.nan
        else:
            return " ".join(directo)
    else:
        cast=[]
        for i in x:
            cast.append(i.get('name'))
        if cast == []:
            return np.nan
        else:
            index=actor_index-1
            if index == 0:
                return cast[index]
            else:
                if len(cast)<actor_index:
                    return np.nan
                else:
                    return (cast[index])
        
                
            
```

```python
df_2017['actor_1_name']=df_2017['cast'].apply(lambda x: get_cast(x,actor_index=1))
df_2017['actor_2_name']=df_2017['cast'].apply(lambda x: get_cast(x,actor_index=2))
df_2017['actor_3_name']=df_2017['cast'].apply(lambda x: get_cast(x,actor_index=3))
df_2017['director_name']=df_2017['crew'].apply(lambda x: get_cast(x,director=True))

```

```python
class createdata:
    def __init__(self,wikilink):
        self.wiki_link=wikilink
        self.tmdb_movie=Movie()
        self.tmdb_movie_link="https://api.themoviedb.org/3/movie/"
    def __get_genre(self, x):
        genres = []
        result = self.tmdb_movie.search(x)
        if not result:
            return np.nan
        movie_id = result[0].id
        response = requests.get(self.tmdb_movie_link+"{}?api_key={}".format(movie_id, tmdb.api_key))
        data_json = response.json()
        gen = data_json["genres"]
        if gen:
            for i in range(len(gen)):
                genres.append(gen[i]['name'])
            return ' '.join(genres)
        else:
            return np.nan
            
    def __get_actor_director(self,x,actor_index=None,director=None):
        if director:
            if " (director)" in x:
                return x.split(" (director)")[0]
            elif " (directors)" in x:
                return x.split(" (directors)")[0]
            else:
                return x.split(" (director/screeenplay)")[0]
        else:
            actors = (x.split("screenplay); ")[-1]).split(", ")  ## doubt
            index=actor_index-1
            if index==0:
                return actors[0]
            else:
                if len(actors)<actor_index:
                    return np.nan
                else:
                    return actors[index];
    def _create_dataframe(self):
        WEB_LINK=self.wiki_link
        source=urllib.request.urlopen(WEB_LINK).read()
        soup=bs.BeautifulSoup(source,'lxml')
        tables=soup.find_all('table',{'class':'wikitable sortable'})
        df1=pd.read_html(str(tables[0]))[0]
        df2=pd.read_html(str(tables[1]))[0]
        df3=pd.read_html(str(tables[2]))[0]
        df4=pd.read_html(str(tables[3]).replace("'1\"\'",'"1"'))[0]
        df=df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
        df.dropna(subset=['Title'],inplace=True)
        df['genres']=df['Title'].map(lambda x: self.__get_genre(str(x)))
        df=df[['Title','Cast and crew','genres']]
        df['director_name']=df['Cast and crew'].apply(lambda x: self.__get_actor_director(x,director=True))
        for i in range(1,4):
            df[f"actor_{i}_name"]=df['Cast and crew'].apply(lambda x: self.__get_actor_director(x,actor_index=i))
            df[f"actor_{i}_name"]=df[f"actor_{i}_name"].replace(np.nan,"unknown")
        df.rename(columns={"Title":"movie_title"},inplace=True)
        required_columns = ['movie_title', 'genres', 'director_name', 'actor_1_name', 'actor_2_name','actor_3_name']
        df = df[required_columns]
        df["movie_title"] = df["movie_title"].str.lower()
        return df
        
        
```
